```python
import h2o
```

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
h2o.init(min_mem_size='1024m')
```

```python
h2o.remove_all()
```

```python
h2o.cluster_info()
```

```python
train = h2o.import_file("D:/Digit Recognizer/train.csv")
```

```python
test = h2o.import_file("D:/Digit Recognizer/test.csv")
```

```python
r = train.runif()
```

```python
train_split = train[r  < 0.8]
```

```python
train_split.shape
```

```python
valid_split = train[r >= 0.8]
```

```python
train_split["label"] = train_split["label"].asfactor()
```

```python
valid_split["label"] = valid_split["label"].asfactor()
```

```python
train_split.types
```

```python
model = H2ODeepLearningEstimator(
        distribution="multinomial",
        activation="RectifierWithDropout",
        hidden=[100,200,100],
        input_dropout_ratio=0.2, 
        sparse=True, 
        l1=1e-5, 
        epochs=100)
```

```python
model.train(
        x= train.names[1:785], 
        y=train.names[0], 
        training_frame=train_split, 
        validation_frame=valid_split)
```

```python
model.params
```

```python
model
```

```python
pred = model.predict(test)
```

```python
pred.head()
```

```python
submit_pred = pred[:,0]
submit_pred.head()
```

```python
submission_dataframe = h2o.H2OFrame({'ImageID': list(range(1, len(test)+1))}).cbind(submit_pred)
```

```python
submission_dataframe.set_name(1,"Label")
```

```python
h2o.h2o.export_file(submission_dataframe,path ="Digit_recognizer_output.csv")
```
